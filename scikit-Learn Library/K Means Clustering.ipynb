```python
# Problem Statement
"""A Mall wants to understand its customer base better to improve 
its marketing strategy. The goal is to cluster customers into
groups based on  their annual income and spending score, so the
marketing team can tailor their campaigns to each cluster's preferences and behaviors."""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
data = {"AnnualIncome": np.random.randint(30000, 100000, 100), 'Spending': np.random.randint(1, 100, 100)}
df = pd.DataFrame(data)
df.head()
```

```python
plt.scatter(df['AnnualIncome'], df['Spending'])
plt.title('Customer Data - Annual income vs spending score')
plt.xlabel('Annual Income')
plt.ylabel('Spending')
plt.show()
```

```python
X = df.values
Kmeans = KMeans(n_clusters = 3, random_state = 42)
df["Cluster"] = Kmeans.fit_predict(X)
```

```python
print(df)
```

```python
plt.scatter(df['AnnualIncome'], df['Spending'], c = df["Cluster"], cmap = "rainbow" )
plt.title('KMeans Clustering - AnnualIncome and Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending')
plt.show()
```

```python
user_input = {"Auser_DfnnualIncome": [75000], "SpendingScore": [70]}
user_df = pd.DataFrame(user_input)
user_cluster = Kmeans.predict(user_df)   
print(f"The user belongs to Cluster {user_cluster[0]}")
```
